# Prepping for training

```python
#features matrix in X and Target Variable(Price) in Y
X = cars_data.drop(['CarName','price'],axis=1)
Y = cars_data['price']
```

```python
Y
```

SINCE WE HAVE TO PREDICT CONTINUOS NUMERICAL DATA, REGRESSION IS THE APPROPRIATE METHOD FOR THIS

```python
#splitting into Train and Test dataset
trainX,testX,trainY,testY=train_test_split(X,Y,test_size=0.20,random_state=20)
```

# Import Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
```

# Training Data on Multiple Regression Algorithms

```python
lr = LinearRegression()
rf = RandomForestRegressor()
xgb = GradientBoostingRegressor()
lr.fit(trainX,trainY)
rf.fit(trainX,trainY)
xgb.fit(trainX,trainY)
```

```python
pred_lf = lr.predict(testX)
pred_rf = rf.predict(testX)
pred_xgb = xgb.predict(testX)
```

# Evaluating the Algorithm

```python
from sklearn import metrics
```

```python
score_lf = metrics.r2_score(testY,pred_lf)
score_rf = metrics.r2_score(testY,pred_rf)
score_xgb = metrics.r2_score(testY,pred_xgb)

MAE_lf = metrics.mean_squared_error(testY,pred_lf)
MAE_rf = metrics.mean_squared_error(testY,pred_rf)
MAE_xgb = metrics.mean_squared_error(testY,pred_xgb)
```

```python
algorithms = pd.DataFrame({
    'Models':['LR','RF','XGB'],
    'R2_SCORE':[score_lf,score_rf,score_xgb],
    'Mean Squared Error':[MAE_lf,MAE_rf,MAE_xgb]
})
```

```python
algorithms
```

```python
sns.barplot(x=algorithms['Models'],y=algorithms['R2_SCORE']).set(title='R2 Score')
```

THE R2 SCORE IS A VERY IMPORTANT METRIC FOR EVALUATING REGRESSION MODELS.

THE HIGHER THE R2 SCORE , THE BETTER THE MODEL AT PREDICTION

IN THIS CASE, THE XGB MODEL HAS THE HIGHEST SCORE



```python
sns.barplot(x=algorithms['Models'],y=algorithms['Mean Squared Error']).set(title='Mean Squared Error')
```

As we can Observe, the XGB Algorithm has the best prediction among the other algorithms
